### Question 3 (1 point)
Each company belongs to a sector (e.g., Consumer Discretionary, Health Care, etc). Output the fraction or percentage of companies that belong to each sector. 

```python
# Count the occurrences of each sector
sector_counts = df_stocks['Industry'].value_counts()
# Calculate the fraction of companies in each sector
fraction_sector = sector_counts / len(df_stocks['Company Name'])

# Convert the fraction to percentage
percentage_of_sector = fraction_sector * 100

# Output the result
print(percentage_of_sector)
```

### Question 4 (1 point)

We learned in class how to add, subtract, multiply, or divide two lists. Each column in a dataframe can be viewed as a list. Divide the two columns "Market Cap" and "P/E" and store the quotient in a new variable called df_earningspotential. Print the minimum, maximum and average value in this new "sliced" dataframe, i.e., df_earningspotential.

*FYI: By dividing the market cap by the P/E ratio, you get the total earnings or net income of the company, which is a key indicator of its profitability.*

```python
df_earningspotential = df_stocks['Market Cap']/df_stocks['P/E']
```

```python
df_earningspotential.describe()
```

```python
print("Average earnings potential:", df_earningspotential.mean())
print("Minimum earnings potential:", df_earningspotential.max())
print("Maximum earnings potential:", df_earningspotential.min())
```
